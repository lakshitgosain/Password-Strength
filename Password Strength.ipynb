```python
#for the words a vector is created....tokenizer assumes that the vectors will be considered for a word...here we dont have a word and works for characters.
```

```python
def word_divide_char(inputs):#takes input and for each and everyword, it makes a character
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X=vectorizer.fit_transform(X['password'])
```
